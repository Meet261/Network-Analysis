```python
# Generation of Random network section 3.1.1, figure 3.1
import random
import networkx as nx
import matplotlib.pyplot as plt
import os

# Generate the graph
G = nx.erdos_renyi_graph(100, 0.2, seed=1)

# Draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=100, node_color='lightblue')
nx.draw_networkx_edges(G, pos, width=0.5, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=8, font_family='sans-serif')
plt.axis('off')

# Save the figure as a PDF
plt.savefig("random_network.pdf", format="pdf")

from google.colab import files
files.download("/content/random_network.pdf")

# Show the figure
plt.show()


```

```python
# Generation of Scale-free network section 3.1.2, figure 3.2
import random
import networkx as nx
import matplotlib.pyplot as plt


# Generate the graph
G = nx.barabasi_albert_graph(100, 2, seed=1)# value of m =2, higher the value of m, every node will have atleast m connections

# Draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=100, node_color='lightblue')
nx.draw_networkx_edges(G, pos, width=0.5, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=8, font_family='sans-serif')
plt.axis('off')



# Save the figure as a PDF
plt.savefig("scale_free_network.pdf", format="pdf")

from google.colab import files
files.download("/content/scale_free_network.pdf")

plt.show()
```

```python
# Generation of Small-world network section 3.1.3, figure 3.3
import random
import networkx as nx
import matplotlib.pyplot as plt


# Generate the graph
G = nx.watts_strogatz_graph(n =100, k = 4, p =0.2,seed =1)# value of k =4, higher the value of k, every node will have atleast m connections to their neighbors

# Draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=100, node_color='lightblue')
nx.draw_networkx_edges(G, pos, width=0.5, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=8, font_family='sans-serif')
plt.axis('off')

print(G.number_of_edges())

# Save the figure as a PDF
plt.savefig("small_world_network.pdf", format="pdf")

from google.colab import files
files.download("/content/small_world_network.pdf")

plt.show()
```
