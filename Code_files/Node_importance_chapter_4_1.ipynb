```python
# Node importance scaller plot in random networks section 4 figure 4.1
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random

# Generate a random graph (Erdős–Rényi model)

G = nx.erdos_renyi_graph(n = 100, p = 0.2, seed=1)  # n = Number of nodes # p = Edge probability

# Degree Centrality

degree_centrality = nx.degree_centrality(G)

# Calculate other centrality measures

closeness_centrality = nx.closeness_centrality(G)

betweenness_centrality = nx.betweenness_centrality(G)

eigenvector_centrality = nx.eigenvector_centrality(G)

# Create a scatter plot of degree centrality versus other centrality measures
plt.figure(figsize=(5, 4))

plt.scatter(list(degree_centrality.values()), list(closeness_centrality.values()), label='Closeness')
plt.scatter(list(degree_centrality.values()), list(betweenness_centrality.values()), label='Betweenness')
plt.scatter(list(degree_centrality.values()), list(eigenvector_centrality.values()), label='Eigenvector')

plt.xlabel('Degree Centrality')
plt.ylabel('Centrality Measures')
plt.title('Node Importance in a Random Graph')
plt.legend()
plt.grid(True)
# To save the figure as a PDF
plt.savefig("random_network_scaller_plot.pdf", format="pdf")

from google.colab import files
files.download("/content/random_network_scaller_plot.pdf")
plt.show()
```

```python
# Node importance correlation matrix in random networks section 4 figure 4.1
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.stats as stats

# Generate a random network (Erdős–Rényi model)

G = nx.erdos_renyi_graph(n = 100, p = 0.2, seed=1)  # n = Number of nodes # p = Edge probability

# Calculate centrality measures
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)

# To calculate Correlations
correlation_degree_closeness = stats.spearmanr(list(degree_centrality.values()), list(closeness_centrality.values())).correlation
correlation_degree_betweenness = stats.spearmanr(list(degree_centrality.values()), list(betweenness_centrality.values())).correlation
correlation_degree_eigenvector = stats.spearmanr(list(degree_centrality.values()), list(eigenvector_centrality.values())).correlation
correlation_closeness_betweenness = stats.spearmanr(list(closeness_centrality.values()), list(betweenness_centrality.values())).correlation
correlation_closeness_eigenvector = stats.spearmanr(list(closeness_centrality.values()), list(eigenvector_centrality.values())).correlation
correlation_betweenness_eigenvector = stats.spearmanr(list(betweenness_centrality.values()), list(eigenvector_centrality.values())).correlation

# To create a Correlation matrix
correlation_matrix = np.array([
    [1.0, correlation_degree_closeness, correlation_degree_betweenness, correlation_degree_eigenvector],
    [correlation_degree_closeness, 1.0, correlation_closeness_betweenness, correlation_closeness_eigenvector],
    [correlation_degree_betweenness, correlation_closeness_betweenness, 1.0, correlation_betweenness_eigenvector],
    [correlation_degree_eigenvector, correlation_closeness_eigenvector, correlation_betweenness_eigenvector, 1.0]
])

# Correlation values
print("Correlation Degree-Closeness:", correlation_degree_closeness)
print("Correlation Degree-Betweenness:", correlation_degree_betweenness)
print("Correlation Degree-Eigenvector:", correlation_degree_eigenvector)
print("Correlation Closeness-Betweenness:", correlation_closeness_betweenness)
print("Correlation Closeness-Eigenvector:", correlation_closeness_eigenvector)
print("Correlation Betweenness-Eigenvector:", correlation_betweenness_eigenvector)

custom_vmin = 0  # Minimum value for color mapping
custom_vmax = 1.0   # Maximum value for color mapping

# Plot the correlation matrix with the custom color range
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest', vmin=custom_vmin, vmax=custom_vmax)
plt.colorbar()
plt.xticks(range(4), ['Degree', 'Closeness', 'Betweenness', 'Eigenvector'], rotation=45, fontsize = 12)
plt.yticks(range(4), ['Degree', 'Closeness', 'Betweenness', 'Eigenvector'], fontsize = 12)
plt.tight_layout()
plt.title('Random Network Correlation Matrix')

# To save the figure as a PDF
plt.savefig("random_network_correlation_matrix.pdf", format="pdf")

from google.colab import files
files.download("/content/random_network_correlation_matrix.pdf")
plt.show()

```
